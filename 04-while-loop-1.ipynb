## ‡∏ó‡∏≥‡∏ã‡πâ‡∏≥‡πÅ‡∏ö‡∏ö‡∏£‡∏π‡πâ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏£‡∏≠‡∏ö ##


1. **‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô (initialize) ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏£‡∏≠‡∏ö (control variable)**
   <br>‡∏ó‡∏≥‡∏Å‡πà‡∏≠‡∏ô loop
2. **‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏Ç‡∏≠‡∏á control variable**
   <br>‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ï‡∏±‡∏î‡∏™‡∏¥‡∏ô‡πÉ‡∏à‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏£‡∏≠‡∏ö‡∏ß‡πà‡∏≤ ‡∏à‡∏∞‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á‡∏†‡∏≤‡∏¢‡πÉ‡∏ô loop ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
3. **update ‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á control variable**
   <br>‡∏ó‡∏≥‡πÉ‡∏ô loop ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏Ñ‡πà‡∏≤ control variable ‡πÉ‡∏´‡∏°‡πà‡πÉ‡∏ô‡∏£‡∏≠‡∏ö‡∏ñ‡∏±‡∏î‡πÑ‡∏õ ‡πÅ‡∏•‡∏∞‡πÉ‡∏´‡πâ‡∏à‡∏ö‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á loop ‡πÑ‡∏î‡πâ


````python
# üìÉ ‡∏û‡∏¥‡∏°‡∏û‡πå‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏ï‡πá‡∏°‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏ï‡πà 1 ‡∏ñ‡∏∂‡∏á 100
i = 1                         # initialize control variable
while i <= 100:   # test condition of control variable
    print(i)
    i = i + 1               # update control variable
````

````python
# üìÉ ‡∏û‡∏¥‡∏°‡∏û‡πå‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏π‡πà‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏ï‡πà 50 ‡∏ñ‡∏∂‡∏á 100
i = 50                     # initialize control variable
while i <= 100:  # test condition of control variable
    if i % 2 == 0:
        print(i)
    i = i + 1              # update control variable
````

````python
# üìÉ ‡∏£‡∏±‡∏ö‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô 5 ‡∏Ñ‡∏ô ‡πÅ‡∏•‡πâ‡∏ß‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏ô
num_students = 5
i = 0               # initalize control variable
while i < num_students:
    # Input
    score = float(input("Enter student's score: "))
    # Process
    if score < 0 or score > 100:           # ‡πÅ‡∏¢‡∏Å‡∏Ç‡∏¢‡∏∞
        print("Score is out of range.")
    else:
        i = i + 1           # ‡∏ô‡∏±‡∏ö‡∏£‡∏≠‡∏ö‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
        if score >= 80:
            grade = "A"
        elif score >= 75:
            grade = "B+"
        elif score >= 70:
            grade = "B"
        elif score >= 65:
            grade = "C+"
        elif score >= 60:
            grade = "C"
        elif score >= 55:
            grade = "D+"
        elif score >= 50:
            grade = "D"
        else: 
            grade = "F"
        # Output
        print("Student {} Score: {:.2f} grade {:s}".format(i, score, grade))
````

````python
# üìÉ ‡∏£‡∏±‡∏ö‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô 5 ‡∏Ñ‡∏ô ‡πÅ‡∏•‡πâ‡∏ß‡∏´‡∏≤‡∏ú‡∏•‡∏£‡∏ß‡∏° ‡πÅ‡∏•‡∏∞‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏ó‡∏±‡πâ‡∏á 5 ‡∏Ñ‡∏ô
num_students = 5
total = 0                 # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏Å‡πá‡∏ö‡∏ú‡∏•‡∏£‡∏ß‡∏°‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô
count = 0               # initalize control variable
while count < num_students:
    # Input
    score = float(input("Enter student's score: "))
    # Process
    if score < 0 or score > 100:           # ‡πÅ‡∏¢‡∏Å‡∏Ç‡∏¢‡∏∞
        print("Score is out of range.")
    else:
        count = count + 1           # ‡∏ô‡∏±‡∏ö‡∏£‡∏≠‡∏ö‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
        total = total + score
        
# Output        
print("Total scores: {:.2f}".format(total))
print("Average score: {:.2f}".format(total / num_students))
````

## 2. ‡∏ó‡∏≥‡∏ã‡πâ‡∏≥‡πÅ‡∏ö‡∏ö‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏£‡∏≠‡∏ö ##


1. **‡πÉ‡∏ä‡πâ `while True`**<br>
    ‡∏ó‡∏≥‡∏ã‡πâ‡∏≥‡πÑ‡∏õ‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢ ‡πÜ ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏£‡∏≠‡∏ö (‡πÅ‡∏ï‡πà‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤‡∏à‡∏∞‡∏™‡∏¥‡πâ‡∏ô‡∏™‡∏∏‡∏î‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÉ‡∏î)
2. **break ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏à‡∏ö‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á loop**<br>
    ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏à‡∏ö‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á loop ‡πÉ‡∏´‡πâ‡πÉ‡∏ä‡πâ `break`

> ‡∏´‡∏≤‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏Ç‡πâ‡∏≤‡∏°‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡∏†‡∏≤‡∏¢‡πÉ‡∏ô loop ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏£‡∏≠‡∏ö‡πÉ‡∏´‡∏°‡πà‡∏ó‡∏±‡∏ô‡∏ó‡∏µ ‡πÉ‡∏´‡πâ‡πÉ‡∏ä‡πâ `continue`

````python
# üìÉ ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤ ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏ï‡πá‡∏°‡∏ó‡∏µ‡πà‡∏£‡∏±‡∏ö‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ ‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏π‡πà‡∏´‡∏£‡∏∑‡∏≠‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏µ‡πà ‡∏à‡∏ô‡∏Å‡∏ß‡πà‡∏≤‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏à‡∏∞‡πÉ‡∏™‡πà -1

count = 0
while True:
    # Input
    number = int(input("Enter number: "))
    # Process
    if number == -1:               # ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏Å‡∏≤‡∏£‡∏à‡∏ö loop
        break
    if number < 0:                   # ‡πÅ‡∏¢‡∏Å‡∏Ç‡∏¢‡∏∞
        print("Invalid number.")
        continue                       # ‡∏Ç‡πâ‡∏≤‡∏°‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠ ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏£‡∏≠‡∏ö‡πÉ‡∏´‡∏°‡πà‡∏ó‡∏±‡∏ô‡∏ó‡∏µ
    count = count + 1
    # Output
    if number % 2 == 0:
        print("{} is even number.".format(number))
    else:
        print("{} is odd number.".format(number))
print("You have checked {} numbers.".format(count))
print("End Program.")
````

````python
# üìÉ ‡∏û‡∏¥‡∏°‡∏û‡πå‡∏™‡∏π‡∏ï‡∏£‡∏Ñ‡∏π‡∏ì ‡∏Ç‡∏≠‡∏á‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏ï‡πá‡∏°‡∏ó‡∏µ‡πà‡∏£‡∏±‡∏ö‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ ‡∏à‡∏ô‡∏Å‡∏ß‡πà‡∏≤‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏à‡∏∞‡πÉ‡∏™‡πà -1

while True:
    # Input 
    number = int(input("Enter number: "))
    if number == -1:              # ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏Å‡∏≤‡∏£‡∏à‡∏ö loop
        break
    if number < 0:                  # ‡πÅ‡∏¢‡∏Å‡∏Ç‡∏¢‡∏∞
        print("Invalid number.")
        continue
    # Nested Loop
    count = 1                           # initialize control variable
    while count <= 12:          # test condition of control variable
        print("{:2d} x {:2d} = {:4d}".format(number, count, number * count))
        count = count + 1         # update control variable
print("End Program.")
````
