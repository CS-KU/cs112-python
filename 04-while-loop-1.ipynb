## ทำซ้ำแบบรู้จำนวนรอบ ##


1. **กำหนดค่าเริ่มต้น (initialize) สำหรับนับจำนวนรอบ (control variable)**
   <br>ทำก่อน loop
2. **ทดสอบเงื่อนไขของ control variable**
   <br>เพื่อตัดสินใจก่อนเริ่มแต่ละรอบว่า จะทำงานคำสั่งภายใน loop หรือไม่
3. **update ค่าของ control variable**
   <br>ทำใน loop เพื่อให้ทดสอบค่า control variable ใหม่ในรอบถัดไป และให้จบการทำงานของ loop ได้


````python
# พิมพ์จำนวนเต็มตั้งแต่ 1 ถึง 100
i = 1                         # initialize control variable
while i <= 100:    # test condition of control variable
    print(i)
    i = i + 1               # update control variable
````

````python
# พิมพ์จำนวนคู่ตั้งแต่ 50 ถึง 100
i = 50                     # initialize control variable
while i <= 100:  # test condition of control variable
    if i % 2 == 0:
        print(i)
    i = i + 1              # update control variable
````

````python
# รับคะแนนของนักเรียน 5 คน แล้วประเมินระดับคะแนนของนักเรียนแต่ละคน
num_students = 5
count = 0               # initalize control variable
while i < num_students:
    # Input
    score = float(input("Enter student's score: "))
    # Process
    if score < 0 or score > 100:           # แยกขยะ
        print("Score is out of range.")
    else:
        i = i + 1           # นับรอบเมื่อคะแนนถูกต้อง
        if score >= 80:
            grade = "A"
        elif score >= 75:
            grade = "B+"
        elif score >= 70:
            grade = "B"
        elif score >= 65:
            grade = "C+"
        elif score >= 60:
            grade = "C"
        elif score >= 55:
            grade = "D+"
        elif score >= 50:
            grade = "D"
        else: 
            grade = "F"
        # Output
        print("Student %d Score: %.2f grade %s" % (i, score, grade))
````

````python
# รับคะแนนของนักเรียน 5 คน แล้วหาผลรวม และค่าเฉลี่ยของคะแนนของนักเรียนทั้ง 5 คน
num_students = 5
total = 0                 # สำหรับเก็บผลรวมคะแนน
count = 0               # initalize control variable
while count < num_students:
    # Input
    score = float(input("Enter student's score: "))
    # Process
    if score < 0 or score > 100:           # แยกขยะ
        print("Score is out of range.")
    else:
        count = count + 1           # นับรอบเมื่อคะแนนถูกต้อง
        total = total + score
        
# Output        
print("Total scores: %.2f" % (total))
print("Average score: %.2f" % (total / num_students))
````

## 2. ทำซ้ำแบบไม่รู้จำนวนรอบ ##


1. **ใช้ while True**<br>
    ทำซ้ำไปเรื่อย ๆ เพราะไม่รู้จำนวนรอบ
2. **break เมื่อต้องการจบการทำงานของ loop**<br>
    เมื่อต้องการจบการทำงานของ loop ให้ใช้ `break`

*หากต้องการข้ามคำสั่งที่เหลือภายใน loop แล้วเริ่มต้นรอบใหม่ทันที ให้ใช้ `continue`

````python
# ตรวจสอบว่า จำนวนเต็มที่รับเข้ามา เป็นจำนวนคู่หรือจำนวนคี่ จนกว่าผู้ใช้จะใส่ -1

count = 0
while True:
    # Input
    number = int(input("Enter number: "))
    # Process
    if number == -1:               # เงื่อนไขการจบ loop
        break
    if number < 0:                   # แยกขยะ
        print("Invalid number.")
        continue                       # ข้ามคำสั่งที่เหลือ แล้วเริ่มรอบใหม่ทันที
    count = count + 1
    # Output
    if number % 2 == 0:
        print("%d is even number." % number)
    else:
        print("%d is odd number." % number)
print("You have checked %d numbers." % count)
print("End Program.")
````

````python
# พิมพ์สูตรคูณ ของจำนวนเต็มที่รับเข้ามา จนกว่าผู้ใช้จะใส่ -1

while True:
    # Input 
    number = int(input("Enter number: "))
    if number == -1:              # เงื่อนไขการจบ loop
        break
    if number < 0:                  # แยกขยะ
        print("Invalid number.")
        continue
    # Nested Loop
    count = 1                           # initialize control variable
    while count <= 12:          # test condition of control variable
        print("%d x %2d = %4d" % (number, count, number * count))
        count = count + 1         # update control variable
print("End Program.")
````
